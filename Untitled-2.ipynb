```python
def encdummy(df, name):
    dummies = pd.get_dummies(df[name])
    for x in dummies.columns:
        dummy_name = f"{name}-{x}"
        df[dummy_name] = dummies[x]
    df.drop(name, axis=1, inplace=True)

encdummy(df,'mainroad')
encdummy(df,'guestroom')
encdummy(df,'basement')
encdummy(df,'hotwaterheating')
encdummy(df,'airconditioning')
encdummy(df,'prefarea')
encdummy(df,'furnishingstatus')

df.head()
```

```python
X = df.drop(columns=["price"])
y = df["price"]

# print(y)
```

```python
class stdscaler:
    def __init__(self):
        self.mean = None
        self.std = None
        
    def fit(self, X):
        self.mean = np.mean(X, axis=0)
        print(self.mean)
        self.std = np.std(X, axis=0)
        print(self.std)
        return self
        
    
    def transform(self, X):
        return (X - self.mean) / self.std
    
    def fittransform(self, X):
        self.fit(X)
        return self.transform(X)

scaler = stdscaler()
X_scaled = scaler.fittransform(X)

ymean = np.mean(y,axis=0)
ystd = np.std(y,axis=0)
y = scaler.fittransform(y)
X_train=X_scaled
```

```python
rate = 5e-4
iterations = 10000
larray = []
m,nx = X_train.shape
W = np.zeros(nx)
b = 0

for i in range(iterations):
  y1 = np.dot(X_train,W) + b
  dz = y1 - y
    # dw = (np.dot(Xtr,dz.T))/m
  dw = (np.dot(X_train.T,dz))/m
  db = np.mean(dz)

  W -= rate * dw
  b -= rate * db

  loss = np.mean((y1-y)**2)
  larray.append(loss)
  print(f"Iteration {i+1}: Loss = {loss:.2e}")
            # if (iterations%1000==0):
            #     rate/=2
            # print(y1)

```

```python
f1 = plt.figure(1)
plt.plot(range(1, iterations + 1), larray)
plt.xlabel('Iterations')
plt.ylabel('Loss')



y1 = np.dot(X_train , W) + b
y1 = ymean+y1*ystd
y = ymean+y*ystd

f2 = plt.figure(2)
plt .scatter(y, y1)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')



# f4 = plt.figure(4)
# plt.scatter(y,xcorr)
# plt.xlabel('Actual Prices')
# plt.ylabel('correlation matrix')
plt.show()
```
